```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def TM_i(n1,n2):
    return np.array([[(n2+n1)/(2*n2), (n2-n1)/(2*n2)], [(n2-n1)/(2*n2), (n2+n1)/(2*n2)]])
def TM_d(n,d,k):
    return np.array([[np.exp(1j*n*k*d),0],[0,np.exp(-1j*n*k*d)]])
```

```python
n1 = 1
n2 = 3.5
n3 = 1
d = 0.5
wavl = 0.2
Tm = TM_i(3.5,1) @ TM_d(3.5,d, 2*np.pi/wavl) @ TM_i(1,3.5)
```

```python
Tc = np.abs(Tm[0,0] - Tm[0,1]*Tm[1,0]/Tm[1,1])
Rc = np.abs(- Tm[1,0]/Tm[1,1])
print(Tc**2, Rc**2)
```

Light enter from material1(index=n1) $\rightarrow$ material2(index=n2) $\rightarrow$ material3(index=n3)

```python
def TRc(n1,n2,n3,d,k):
    Tm = TM_i(n2,n3) @ TM_d(n2,d,k) @ TM_i(n1,n2)
    Tc = np.abs(Tm[0,0] - Tm[0,1]*Tm[1,0]/Tm[1,1])
    Rc = np.abs(- Tm[1,0]/Tm[1,1])
    return Tc**2, Rc**2
```

```python
wvl_inv = np.linspace(0,4,1000)
Tcl=[]
Rcl=[]
for wvi in wvl_inv:
    Tc, Rc = TRc(1,3.5,1,0.5,2*np.pi*wvi)
    Tcl.append(Tc)
    Rcl.append(Rc)
plt.plot(wvl_inv, Tcl)
plt.plot(wvl_inv, Rcl)
```

```python
def Tr_M(nl, dl, k): #nl is index array, dl is thickness array, k is the wave vector of incoming wave.
    TI = np.array([[(nl[0]+1)/(2*nl[0]), (nl[0]-1)/(2*nl[0])], [(nl[0]-1)/(2*nl[0]), (nl[0]+1)/(2*nl[0])]])
    TD = np.array([[np.exp(1j*nl[0]*k*dl[0]),0],[0,np.exp(-1j*nl[0]*k*dl[0])]])
    TM = TD @ TI
    for i, n in enumerate(nl):
        if (i+1) < len(nl):            
            n1 = n
            n2 = nl[i+1]
            d = dl[i+1]
            TI = np.array([[(n2+n1)/(2*n2), (n2-n1)/(2*n2)], [(n2-n1)/(2*n2), (n2+n1)/(2*n2)]])
            TD = np.array([[np.exp(1j*n2*k*d),0],[0,np.exp(-1j*n2*k*d)]])
            TM = (TD @ TI) @ TM
    n1 = nl[-1]
    n2 = 1
    TI = np.array([[(n2+n1)/(2*n2), (n2-n1)/(2*n2)], [(n2-n1)/(2*n2), (n2+n1)/(2*n2)]])
    TM = TI @ TM
    return -TM[1,0]/TM[1,1], TM[0,0] - TM[0,1]*TM[1,0]/TM[1,1]
```

```python
l0 = 1.55
nl = [3,1,3,1,3,1,3]
dl = [l0/4/3,l0/4,l0/4/3,l0/4,l0/4/3,l0/4,l0/4/3]
```

```python
fl = np.linspace(0.01,4,500)
Tcl = []
Rcl = []
for f in fl:
    Rc, Tc = Tr_M(nl,dl,2*np.pi*f)
    Tcl.append(np.abs(Tc)**2)
    Rcl.append(np.abs(Rc)**2)
#plt.plot(fl,Tcl)
plt.xlim(0,4)
plt.plot(fl,Tcl)
```

```python
plt.xlim(0,4)
plt.plot(fl,Rcl)
```
